```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# Get the shape of the first image in the dataset
first_image = train_images[0]
height, width = first_image.shape

# Desired cropped dimensions
crop_height = 20
crop_width = 20

# Calculate vertical space
vertical_space = height - crop_height
print(f"Vertical space (height - crop height): {vertical_space} pixels")

# Calculate starting y-coordinate (vertical starting point)
start_y = vertical_space // 2
print(f"Starting y-coordinate after dividing by 2: {start_y} pixels")

# Calculate horizontal space
horizontal_space = width - crop_width
print(f"Horizontal space (width - crop width): {horizontal_space} pixels")

# Calculate starting x-coordinate (horizontal starting point)
start_x = horizontal_space // 2
print(f"Starting x-coordinate after dividing by 2: {start_x} pixels")

# Calculate ending coordinates
end_y = start_y + crop_height
end_x = start_x + crop_width
print(f"Ending y-coordinate: {end_y} pixels")
print(f"Ending x-coordinate: {end_x} pixels")

# Crop the image
cropped_img = first_image[start_y:end_y, start_x:end_x]

# Display the original and cropped images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(first_image, cmap='gray')
plt.axhline(y=start_y, color='red', linestyle='--', label=f'Start_y = {start_y}')
plt.axhline(y=end_y, color='green', linestyle='--', label=f'End_y = {end_y}')
plt.axvline(x=start_x, color='blue', linestyle='--', label=f'Start_x = {start_x}')
plt.axvline(x=end_x, color='purple', linestyle='--', label=f'End_x = {end_x}')
plt.title("Original Image with Indicated Crop Start and End Points")
plt.axis('off')
plt.legend()

plt.subplot(1, 2, 2)
plt.imshow(cropped_img, cmap='gray')
plt.title("Cropped Image")
plt.axis('off')

plt.show()
```
